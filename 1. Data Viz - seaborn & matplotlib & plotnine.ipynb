```python
# import packages
import os 
import pandas as pd
import seaborn as sns
import plotnine
from plotnine import ggplot, ggtitle, scale_y_log10, aes, geom_line, geom_point, scale_alpha, ylim, xlim, scale_x_log10, scale_size, labs, guides, scale_size_continuous, theme, scale_color_discrete
# os.getcwd()
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

### Task 1 
1. Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

x = income, y = life expectancy, size = population, color = four_regions

```python
# 2.  Import the q1data.csv dataset.
q1 = pd.read_csv("q1data.csv")
q1.head()
```

```python
# 3. Recreate the plot as best you can using the library of your choice (e.g. matplotlib, seaborn, altair, plotly, etc.).
# using plotnine package
(
    ggplot(data = q1) 
    + aes(x="income", y = "life_exp")
    + geom_point(aes(size = "population", color = "four_regions"))  
    + ylim(10, 90)  
    + xlim(10, 128000) 
    + scale_x_log10()
    + labs(y = "Life Expectancy", x = "Log(income)")
    + ggtitle("Life Expectancy vs Log of income")
)
```

### Task 2
1. Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.
x = exports, y = imports, hue = four_regions, size = energy

```python
# 2. Import the q2data.csv dataset.
q2 = pd.read_csv("q2data.csv")
```

```python
# 3. Recreate the plot as best you can using the library of your choice (e.g. matplotlib, seaborn, altair, plotly, etc.).
# using seaborn and matplotlib
sns.set(rc = {'figure.figsize':(15,8)})
p = sns.scatterplot(data=q2, x="exports", y="imports", size="energy", legend=False, hue = "four_regions", sizes=(150, 400))
p.set_xlabel("Exports (\% of GDP)", fontsize = 16)
p.set_ylabel("Imports (\% of GDP)", fontsize = 16)

# control x and y limits
plt.ylim(0, 400)
plt.xlim(0, 230)

plt.show()
```

### Task 3 
1. Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.
x = internet_users, y = gdp, color = four_regions, size = gdp

```python
# 2. Import the q3data.csv dataset
q3 = pd.read_csv("q3data.csv")
q3.head()
```

```python
# 3. Recreate the plot as best you can using the library of your choice (e.g. matplotlib, seaborn, altair, plotly, etc.).
# using plotnine package
(
    ggplot(data = q3) 
    + aes(x = "internet_users", y = "gdp")
    + geom_point(aes(color = "four_regions", size = "income"), show_legend = False)
    + labs(x = "Individuals using the internet", y = "GDP/capita")  
    + scale_y_log10(guide=False)
    + ggtitle("GDP per Capita vs Internet users")
)
```
